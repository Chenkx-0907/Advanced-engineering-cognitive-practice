# using FFT to fit the experiment data

## Import the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Experiment 2
### Import data

```python
# Load the xlsx file into a pandas ExcelFile object
xls = pd.ExcelFile('./data/exp2/exp2.xlsx')

# Create a dictionary to store the sheets as DataFrames
sheets = {}
for sheet_name in xls.sheet_names:
    sheets[sheet_name] = xls.parse(sheet_name)

# Convert the DataFrames to NumPy arrays and store them in a dictionary
arrays = {}
for sheet_name, sheet_df in sheets.items():
    arrays[sheet_name] = np.array(sheet_df)
```

### Plot

```python
# Create a figure and axis object
fig, ax1 = plt.subplots(figsize = (18,6), dpi=100)

# Show Chinese
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

# conform the x and y
t = arrays['0'][:,0]
y = arrays['0'][:,1]

# Plot the data
ax1.plot(t, y)
ax1.set_xlabel('time (s)',size=14)
ax1.set_ylabel('Amplitude',size=14)
ax1.set_title('',size=20)

# Display the plot
plt.show()
```

### FFT Result

```python
# Compute the FFT of the signal
N = len(y)
Fs = N/max(t)                               # sampling frequency
freq = np.abs(np.fft.fftfreq(N, d=1/Fs))    # frequency axis in Hz
y_fft = np.abs(np.fft.fft(y))               # FFT Result
y_fft_nom = np.abs(np.fft.fft(y)/N)         # Normalization of the FFT Result

# Slicing
freq = freq[:800]
y_fft = y_fft[:800]
y_fft_nom = y_fft_nom[:800]

# Find the max magnitude frequency
max_index = np.argmax(y_fft)    # max index
max_value = freq[max_index]     # max magnitude value

# Plot the FFT result
fig, ax2 = plt.subplots(figsize=(18,6), dpi=100)
ax2.plot(freq,y_fft)
ax2.set_xlabel('Frequency (Hz)',size=14)
ax2.set_ylabel('Magnitude',size=14)
ax2.set_title('FFT Result',size=20)

# Plot the max magnitude frequency
ax2.plot(max_value,y_fft[max_index],'ks')
show_max='  '+f'{max_value:.2f}'+' HZ'
ax2.annotate(show_max,xy=(max_value,y_fft[max_index]),xytext=(max_value,y_fft[max_index]),size=14)

# Display the plot
plt.show()
```
